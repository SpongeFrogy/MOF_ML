```python
DB['folder number'] = DB['folder number'].astype(int)
DB_from['folder number'] = DB_from['folder number'].astype(int)
DB_to['folder number'] = DB_to['folder number'].astype(int)
```

```python
DB.sort_values(by=['folder number'], inplace=True)
DB_from.sort_values(by=['folder number'], inplace=True)
DB_to.sort_values(by=['folder number'], inplace=True)
```

```python
"""
Для лучшего понимания ,на русском.
Этот скрипт принимает 3 базы данных.
Первые две: до и после фазового превращения. 
Третья: по каому правилу (обратимо или нет) и под каким стимулом они переходят
Каждая БД должна иметь колонку-ключ в первом столбце. Это порядковый номер в Гугл-таблице
First_DB == DB_from т.е. до превращения 

final_DB == датасет с первой колонкой-ключом ('folder number'), далее - признаки, после колонка - стимулы по которым происходит переход
Фильтрация последней колонки будет сделана в другой функции
"""
def create_new_DB_from_2 (first_DB:pd.DataFrame, second_DB:pd.DataFrame, target_DB:pd.DataFrame, key_column_name:str):
    # change target_DB to numpy array
    target_DB_np = target_DB.to_numpy()

    # create new column in first_DB and second_DB by special rule
    for folder_number , revers, stimuli in target_DB_np: 
        
        first_DB.loc[first_DB[key_column_name] == folder_number, 'Stimuli'] = stimuli

        if revers == 'yes':
            second_DB.loc[second_DB[key_column_name] == folder_number, 'Stimuli'] = stimuli
    
    # unite two databases
    final_DB = pd.concat([first_DB, second_DB], ignore_index=True)

    # drop rows with stimuli == NaN
    final_DB.dropna(subset=['Stimuli'], inplace=True)
    
    
    return final_DB
```

```python
test_def = create_new_DB_from_2(DB_from, DB_to, DB, 'folder number')
```
