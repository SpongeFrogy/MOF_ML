```python
import pandas as pd
```

```python
# Read all .csv files 

SG_DB_from = pd.read_csv('SG_dataset_from.csv')
SG_DB_to = pd.read_csv('SG_dataset_to.csv')
SG_DB_unmarked = pd.read_csv('SG_dataset_unmarked.csv')

DB_google = pd.read_csv('Database_from_google.csv', delimiter=';')

DB_main = pd.read_csv('main_dataset_from_2_databases_preprocessed.csv')
```

```python
# Concat all DBs with SG's
one_DB = pd.concat([SG_DB_from,SG_DB_to, SG_DB_unmarked])
```

```python
one_DB.reset_index()
```

```python
# Rename column name as in one_DB and add postfix '.cif' to file name column 
DB_main.rename(columns={'file_name': 'file name'}, inplace=True)
DB_main['file name'] = DB_main['file name'] + '.cif'
```

```python
DB_main.head()
```

```python
# Merge one_DB with main_DB by file_name

DB_merge = pd.merge(DB_main, one_DB, on='file name')
```
