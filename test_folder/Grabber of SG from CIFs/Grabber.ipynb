```python

import os, fnmatch
from gemmi import cif
import pandas as pd
```

```python
def table_to_pdRow(tab, n=7):
    if tab.width() > 6:
        pdrow = [cif.as_number(v) for v in tab[0]]
        pdrow[-1] = tab[0][-1]
        return [pdrow]
```

```python
def cif_SG(path):
#     df = pd.DataFrame(columns=["a","b","c",
#                  "alpha","beta","gamma",
#                  "symmetry", 'SG_Hall', "SG_IT_number" ])
#     columns = df.columns
    print(main_folder+ '\\' + path)
    items = ["_cell_length_a", "_cell_length_b", "_cell_length_c",
            "_cell_angle_alpha", "_cell_angle_beta", "_cell_angle_gamma ",'_space_group_crystal_system'
            ,'_space_group_IT_number' ,"_space_group_name_H-M_alt" ,'_space_group_name_Hall']
    try:
        new_row = []
        for item in items: 
            value = cif.read_file(main_folder+ '\\'+ f'{path}').sole_block().find_values(item)
            if (value.__len__() != 0):
                new_row.append(value.__getitem__(0))
            else:
                new_row.append('None')
                
    except ValueError:
        print('Something wrong with file')
    
    return new_row
```

```python
def try_get_SG(path, folder):
    
    identifiers = cif_SG(path)
    if folder[0] == '(' and folder[2] != ')':
        # print(folder[1:3] +' '+ folder)
        identifiers.append(folder[1:3])
    elif folder[0] != '(' and folder[2] == ')':
        # print(folder[0:2] +' '+ folder)
        identifiers.append(folder[0:2])
    elif folder[2] == ')':
        # print(folder[1:2] +' '+ folder)
        identifiers.append(folder[1:2])
    # print(identifiers)
                
    
    
    return identifiers
```

```python
main_folder = f"D:\github repo\MOF_ML\\test_folder\Grabber of SG from CIFs\cifs"

if __name__ == '__main__':
    DB1 = []
    DB2 = []
    DB3 = []
    for folder in os.listdir(main_folder):
        for file in os.listdir(f"{main_folder}/{folder}"):
            if fnmatch.fnmatch(file, '*.cif'):
                if file.lower().find('from') != -1 or file.lower().find('initial') != -1 or file.lower().find('phase i') != -1 or file.lower().find('based on') != -1:
                    x1 = try_get_SG(f"{folder}\{file}",folder)
                    x1.append(file)
                    DB1.append(x1)
                elif file.lower().find('to') != -1 or file.lower().find('phase ii') != -1:
                    x2 = try_get_SG(f"{folder}\{file}",folder)
                    x2.append(file)
                    DB2.append(x2)
                else:
                    x3= try_get_SG(f"{folder}\{file}",folder)
                    x3.append(file)
                    DB3.append(x3)
    

    
    
```

```python
# Save to csv

df1 = pd.DataFrame(DB1, columns=["a","b","c",'alpha','beta','gamma','crystal_system'
            ,'SG_IT_number' ,"SG_H-M_alt" ,'SG_Hall', 'folder number','file name'])
    
df1.to_csv("D:\github repo\MOF_ML\\test_folder\Grabber of SG from CIFs\SG_dataset_from.csv", index=False)

df2 = pd.DataFrame(DB2, columns=["a","b","c",'alpha','beta','gamma','crystal_system'
            ,'SG_IT_number' ,"SG_H-M_alt" ,'SG_Hall', 'folder number','file name'])
    
df2.to_csv("D:\github repo\MOF_ML\\test_folder\Grabber of SG from CIFs\SG_dataset_to.csv", index=False)

df3 = pd.DataFrame(DB3, columns=["a","b","c",'alpha','beta','gamma','crystal_system'
            ,'SG_IT_number' ,"SG_H-M_alt" ,'SG_Hall', 'folder number','file name'])
    
df3.to_csv("D:\github repo\MOF_ML\\test_folder\Grabber of SG from CIFs\SG_dataset_unmarked.csv", index=False)
```
