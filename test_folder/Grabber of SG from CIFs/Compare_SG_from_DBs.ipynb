```python
import pandas as pd
```

```python
# Read the CSV file with the SGs from the CIFs

DB_from = pd.read_csv('SG_dataset_from.csv')
DB_to = pd.read_csv('SG_dataset_to.csv')
DB_unmarked = pd.read_csv('SG_dataset_unmarked.csv')
```

```python
DB_from.head()
```

```python
# Change 'None' in SG_IT_number to 0
DB_from['SG_IT_number'] = DB_from['SG_IT_number'].replace('None', 0)
DB_to['SG_IT_number'] = DB_to['SG_IT_number'].replace('None', 0)
```

```python
# Sort the DBs by folder number

DB_from['SG_IT_number'] = DB_from['SG_IT_number'].astype(int)
DB_to['SG_IT_number'] = DB_to['SG_IT_number'].astype(int)

DB_from['folder number'] = DB_from['folder number'].astype(int)
DB_to['folder number'] = DB_to['folder number'].astype(int)

DB_from.sort_values(by=['folder number'], inplace=True)
DB_to.sort_values(by=['folder number'], inplace=True)


```

```python
# Take folder numbers from the DBs

folder_numbers_from = DB_from['folder number'].tolist()
folder_numbers_to = DB_to['folder number'].tolist()

# Find the folder numbers that are in DB_from and in DB_to

folder_numbers_common = list(set(folder_numbers_from) & set(folder_numbers_to))
folder_numbers_common

# Find the folder numbers that are in DB_from but not in DB_to

folder_numbers_from_only = list(set(folder_numbers_from) - set(folder_numbers_to))
folder_numbers_from_only

# Find the folder numbers that are in DB_to but not in DB_from

folder_numbers_to_only = list(set(folder_numbers_to) - set(folder_numbers_from))
folder_numbers_to_only
```

```python
# Exctact SG_IT_numbers from DB_from and DB_to for the common folder numbers

DB_from_common = DB_from[DB_from['folder number'].isin(folder_numbers_common)]
DB_to_common = DB_to[DB_to['folder number'].isin(folder_numbers_common)]

DB_from_common.head()
```

```python
DB_to_common_cleaned = DB_to_common[['SG_IT_number','folder number']].drop_duplicates().reset_index(drop=True)
```

```python
type(DB_to_common_cleaned)
```

```python
DB_from_common_cleaned = DB_from_common[['SG_IT_number','folder number']].drop_duplicates().reset_index(drop=True)
```

```python
DB_from_common_cleaned
```

```python
# concat the two DBs

DB_common = pd.merge(DB_from_common_cleaned, DB_to_common_cleaned, on='folder number')
DB_common
```

```python
# Drop the rows where SG_IT_number_x = SG_IT_number_y = 0

DB_common = DB_common[(DB_common['SG_IT_number_x'] != 0) & (DB_common['SG_IT_number_y'] != 0)]
DB_common
```

```python
# Add new column if SG_IT_number_x < SG_IT_number_y

DB_common['lower SG'] = DB_common['SG_IT_number_x'] > DB_common['SG_IT_number_y']
DB_common
```
