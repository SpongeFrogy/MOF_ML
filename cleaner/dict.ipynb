```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('5k_with_symmetry.csv', sep=',', index_col=0)
```

```python
data
```

```python
def columnToDict(dataframe, column_name):
    series = dataframe[column_name]
    uValues = series.unique()
    valueIndex = [None]*len(uValues)
    for i, val in enumerate(uValues):
        valueIndex[i] =  str(series[series == val].index[0])
    return dict(zip(valueIndex, uValues))
```

```python
def numberDict(dataframe, column_name):
    return dict(zip(list(range(len(dataframe[column_name].unique()))), dataframe[column_name].unique()))
```

```python
for name in data.columns:

    d = numberDict(data, name)

    with open(name+ '_number' +'.csv', 'w') as f:
        for key in d.keys():
            f.write("%s, %s\n" % (key, d[key]))
```

```python
for name in data.columns:

    d = columnToDict(data, name)

    with open(name+'.csv', 'w') as f:
        for key in d.keys():
            f.write("%s, %s\n" % (key, d[key]))
```
